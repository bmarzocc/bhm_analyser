# Importing the text files
---
uHTR data which is kept in text format is parsed using the parse_text_file(...) --> kept in tools.parser



```python
#Load uHTR4
ch4, ampl4, tdc4,tdc4_2, bx4,orbit4,run4  = parser.parse_text_file("./data/2022_10_31_QIE_Alignment/uHTR4.txt")
#Load uHTR11
ch11,ampl11, tdc11,tdc11_2, bx11,orbit11,run11  = parser.parse_text_file("./data/2022_10_31_QIE_Alignment/uHTR11.txt")
```

Since loading the uHTR file takes time; we keep the values in variables, and then load it to the bhm_analyser from variables.

Alternatively, one can use bhm_analyser.load_data(self,folder_name)

```python
uHTR4 = bhm_analyser(uHTR="4")
# uHTR4.load_data("./data/2022_10_26_HV_Tests/uHTR4.txt")
uHTR4.load_data_from_memory(ch4,ampl4, tdc4,tdc4_2, bx4,orbit4,run4)


uHTR11 = bhm_analyser(uHTR="11")
# uHTR11.load_data("./data/2022_10_26_HV_Tests/uHTR11.txt")
uHTR11.load_data_from_memory(ch11,ampl11, tdc11,tdc11_2, bx11,orbit11,run11)
```
